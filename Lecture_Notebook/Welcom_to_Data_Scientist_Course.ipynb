### üìä Step 4: ÂèØËßÜÂåñËØÑÂàÜÂàÜÂ∏ÉÔºàVisualize DistributionÔºâ

````python
import matplotlib.pyplot as plt

# Assuming 'Year' is a column in df_Netflix_Dataset_Movie and it represents the year of the movie.
# If the year information is in df_user_rating_movie and is named differently, adjust the column name.
# We group by the year and count the number of unique Movie_IDs for each year.
movies_per_year = df_Netflix_Dataset_Movie.groupby('Year')['Movie_ID'].nunique()

# Plotting the results
plt.figure(figsize=(15, 7))
movies_per_year.plot(kind='bar')
plt.title('Number of Movies per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
# Group by movie name and count the number of user ratings for each movie
movie_ratings_count = df_user_rating_movie.groupby('Name')['User_ID'].count()

# Sort the counts for better visualization (optional, but often helpful)
movie_ratings_count = movie_ratings_count.sort_values(ascending=False)

# Plotting the results
plt.figure(figsize=(20, 8)) # Adjust figure size as needed
movie_ratings_count[:10].plot(kind='bar')
plt.title('Top 10 Number of Ratings per Movie')
plt.xlabel('Movie Name')
plt.ylabel('Number of Users Rated')
plt.xticks(rotation=90, ha='right') # Rotate x-axis labels and align them
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

````python
import matplotlib.pyplot as plt
# Group by movie name and count the number of user ratings for each movie
movie_ratings_count = df_user_rating_movie.groupby('Name')['User_ID'].count()

# Get the top 10 movies with the most ratings
top_10_movies = movie_ratings_count.sort_values(ascending=False).head(10)

# Plotting the top 10 results
plt.figure(figsize=(12, 7)) # Adjust figure size as needed
top_10_movies.plot(kind='bar')
plt.title('Top 10 Movies with Most Ratings')
plt.xlabel('Movie Name')
plt.ylabel('Number of Users Rated')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels and align them for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

````python
import matplotlib.pyplot as plt
# Group by movie name and calculate the average rating for each movie
movie_avg_ratings = df_user_rating_movie.groupby('Name')['Rating'].mean()

# Get the top 10 movies based on their average rating
top_10_highest_rated_movies = movie_avg_ratings.sort_values(ascending=False).head(10)

# Plotting the top 10 highest rated movies
plt.figure(figsize=(12, 7)) # Adjust figure size as needed
top_10_highest_rated_movies.plot(kind='bar')
plt.title('Top 10 Highest Rated Movies')
plt.xlabel('Movie Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels and align them for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

````python
import matplotlib.pyplot as plt
# Get the bottom 10 movies based on their average rating
top_10_lowest_rated_movies = movie_avg_ratings.sort_values(ascending=True).head(10)

# Plotting the bottom 10 lowest rated movies
plt.figure(figsize=(12, 7)) # Adjust figure size as needed
top_10_lowest_rated_movies.plot(kind='bar')
plt.title('Top 10 Lowest Rated Movies')
plt.xlabel('Movie Name')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels and align them for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
# Convert 'Year' to numeric, coercing errors to NaN
df_user_rating_movie['Year'] = pd.to_numeric(df_user_rating_movie['Year'], errors='coerce')

# Drop rows where 'Year' is NaN
df_user_rating_movie.dropna(subset=['Year'], inplace=True)

# Convert 'Year' to integer type
df_user_rating_movie['Year'] = df_user_rating_movie['Year'].astype(int)

# Find the most recent year in the dataset
latest_year = df_user_rating_movie['Year'].max()

# Calculate the start year for the last 10 years
start_year = latest_year - 9

# Filter data for the last 10 years
df_recent_years = df_user_rating_movie[df_user_rating_movie['Year'] >= start_year]

# Group by Year and calculate the average rating
avg_rating_by_year = df_recent_years.groupby('Year')['Rating'].mean()

# Sort by year to ensure correct plotting order
avg_rating_by_year = avg_rating_by_year.sort_index()

# Plotting the average rating for the most recent 10 years
plt.figure(figsize=(12, 6))
avg_rating_by_year.plot(kind='line', marker='o')
plt.title(f'Average Movie Rating in the Last 10 Years ({start_year}-{latest_year})')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.xticks(avg_rating_by_year.index) # Set x-axis ticks to the years
plt.grid(True)
plt.tight_layout()
plt.show()
````
