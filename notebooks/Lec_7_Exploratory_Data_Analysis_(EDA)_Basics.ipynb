# Lesson_07 探索性数据分析 (EDA)


EDA 是数据科学流程中的一个关键的早期步骤，通常发生在数据收集和清洗之后，模型构建之前。会应用一组用于探索、描述和总结数据性质的统计技术，以确保分析的客观性和互操作性。

它的作用是作为一个“侦察兵”，帮助数据科学家在正式建模之前对数据有一个全面的认识，避免盲目建模。EDA 的结果会直接影响后续的数据预处理、特征工程、模型选择和评估。

### EDA 的主要目标

*   发现数据中的模式和趋势。
*   识别和理解变量之间的关系。
*   检测数据中的异常值和离群点。
*   评估数据的质量和完整性（如缺失值）。
*   为后续的统计建模或机器学习任务提供有价值的见解和假设。

### 常用的 EDA 技术和工具

常用的 EDA 技术包括：

*   **可视化：** 绘制直方图、散点图、箱线图、条形图等来探索数据分布和变量关系。
*   **统计摘要：** 计算均值、中位数、标准差、分位数等描述性统计量来总结数据。
*   **分组分析：** 按不同的类别对数据进行分组，比较各组之间的差异。
*   **相关性分析：** 计算变量之间的相关系数，了解它们之间的线性关系。

常用的工具包括 Python 中的 `pandas`、`matplotlib`、`seaborn`，以及 R 等。

### 不同类型的 EDA

#### **1. 单变量分析（Univariate Analysis）：**

单变量分析是分析数据的最简单形式。正如名称“uni”所暗示的那样，单变量分析是对一个变量的分析。它有助于数据分析师了解单个变量的值是如何分布的。它分别探索数据集中的每个变量。它会分析每个变量的特征，但不会研究和其他变量直接的影响或关系。

- 单变量非图形：集中趋势、偏差、偏度、峰度等。
- 单变量图形：直方图、箱线图、饼图、密度图等。

#### **2. 双变量分析（Bivariate Analysis）：**

双变量分析会分析两个变量，并确定它们之间的关系。其中一个变量是因变量，另一个是自变量。它还可以帮助检验简单的关联假设。

双变量非图形：相关分析、回归分析等。
双变量图形：散点图、折线图等。

#### **3. 多变量分析（Multivariate Analysis）：**

当必须同时分析两个以上的变量时，我们就需要进行多变量分析。在处理现实世界的数据时，我们所做的很多预测都是基于多变量数据，因为在给定时间点或一段时间内可能有多个因素影响变量。多元分析的常见类型包括聚类分析、因子分析和多元回归分析。

### 讨论缺失值处理方法

根据之前的分析，数据集在以下列存在缺失值：

*   **Age (年龄):** 缺失比例约为 19.87%。
*   **Cabin (船舱号):** 缺失比例高达 77.10%。
*   **Embarked (登船港口):** 缺失比例约为 0.22%。

其中，'Cabin' 列的缺失值比例非常高，'Age' 列的缺失值比例接近 20%，而 'Embarked' 列的缺失值数量非常少。针对这些存在缺失值的列，我们可以考虑以下几种常见的处理策略：

1.  **删除包含缺失值的行:**
    *   **优点:** 简单直接，不会引入偏差。
    *   **缺点:** 如果缺失值较多，会导致大量数据丢失，尤其是对于像 'Cabin' 这样缺失比例高的列。可能丢失有价值的信息，影响模型的泛化能力。
    *   **适用场景:** 缺失值数量非常少，且删除后对整体数据集影响不大时；或者当缺失是随机发生的 (MCAR - Missing Completely At Random)。

2.  **删除包含缺失值的列:**
    *   **优点:** 简单直接，处理缺失比例极高的列时可以考虑。
    *   **缺点:** 丢失整个列的信息，即使该列可能包含一些有用的信息。
    *   **适用场景:** 缺失值比例非常高（如 'Cabin'），且该列对分析或建模的贡献预期不大时。

3.  **使用统计量填充 (均值、中位数、众数):**
    *   **优点:** 简单易行，保留了所有数据。对于数值型数据，可以使用均值或中位数填充；对于类别型数据，可以使用众数填充。
    *   **缺点:** 引入偏差，尤其是当缺失不是随机发生时。均值填充会影响数据的分布和方差；众数填充会夸大最常见类别的频率。这种方法没有利用到其他列的信息来预测缺失值。
    *   **适用场景:** 缺失值比例不高，且数据近似符合正态分布（均值填充）；数据存在离群值（中位数填充）；类别型数据（众数填充）。适用于缺失是随机发生 (MAR - Missing At Random) 或完全随机发生 (MCAR) 的情况。

4.  **使用模型预测填充:**
    *   **优点:** 可以利用数据集中其他列的信息来预测缺失值，可能比简单统计量填充更准确。
    *   **缺点:** 实现更复杂，需要构建预测模型。如果用于预测的列本身也存在缺失，则需要先处理这些缺失。引入模型误差。
    *   **适用场景:** 缺失值比例适中，且认为缺失值与其他变量存在一定关系 (MAR)。

5.  **基于业务知识或特定规则填充:**
    *   **优点:** 可以结合实际情况和领域知识进行更合理的填充。
    *   **缺点:** 需要深入的业务理解。
    *   **适用场景:** 当对数据的生成机制有清晰的了解，可以推断出合理的缺失值。

### 针对具体列的处理策略讨论

*   **Age (年龄):** 缺失比例约为 20%。删除包含缺失值的行会丢失约 20% 的数据，损失较大。使用均值或中位数填充是常见的简单方法，可以保留数据量。年龄与生存率可能有关联，简单填充可能会引入偏差。更复杂的方法是使用其他特征（如 Pclass, Sex, SibSp, Parch）来预测年龄。考虑到年龄是数值型数据且与生存可能相关，**使用中位数填充**可能是比均值更好的选择，因为中位数对异常值不敏感。
*   **Cabin (船舱号):** 缺失比例高达 77.10%。删除该列是可行的选择，因为它缺失了大部分信息，直接填充会引入大量虚假数据。或者，可以将 'Cabin' 列转换为一个表示“是否有船舱信息”的二元特征（例如，1表示有船舱号，0表示缺失），这可以保留部分信息。**考虑到极高的缺失率，将其转换为二元特征或直接删除**是比较合理的策略。
*   **Embarked (登船港口):** 缺失数量只有 2 个，比例非常低。删除这两行数据对整体数据集影响微乎其微。或者，可以使用众数填充，因为它是类别型数据。**由于缺失值数量极少，直接删除这两行或使用众数填充**都是简单且有效的方法。考虑到众数填充保留了所有数据，可能是稍微更好的选择。

基于上述讨论，在接下来的步骤中，我们将选择以下策略进行处理：
*   **Age:** 使用中位数填充。
*   **Cabin:** 将其转换为一个二元特征（是否有船舱信息）。
*   **Embarked:** 使用众数填充。

```python
# 6. 基于对数据集和业务背景的初步理解，选择一种或多种合适的策略来处理 'Age', 'Cabin', 和 'Embarked' 列的缺失值
# Age: 使用中位数填充
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

# Cabin: 将其转换为一个二元特征（是否有船舱信息）
# 创建新列 'Has_Cabin'，有值则为 1，缺失则为 0
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)
# 可以选择删除原始的 Cabin 列，因为它缺失太多，但为了保留原始数据，暂时不删除，只创建新特征。

# Embarked: 使用众数填充
# 找到 Embarked 列的众数
most_frequent_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_frequent_embarked, inplace=True)

# 7. 再次检查处理后的数据集，确认缺失值已被妥善处理。
print("处理缺失值后的各列缺失值数量:")
print(df.isnull().sum())
```

```python
print("\n处理缺失值后的各列缺失值比例 (%):")
print((df.isnull().sum() / len(df)) * 100)
```

### 识别和讨论异常值

通过上述箱线图，我们可以直观地观察到数据集中一些数值型列的分布以及潜在的异常值（通常表示为箱线图之外的点）。

**观察到的异常值：**

*   **Age (年龄):** 大部分年龄分布在 20 到 40 岁之间，箱线图上方显示有一些高于 65 岁的点，这些可能是年龄较大的乘客。虽然相对较少，但这些点在箱线图上被标记为异常值。
*   **Fare (船票价格):** 绝大多数船票价格较低，集中在 0 到 50 之间。箱线图显示有许多价格非常高的点，有些甚至超过 500。这些极端高价的船票很可能代表头等舱或高等级船舱的乘客，与大多数乘客的价格差异巨大。
*   **SibSp (同乘的兄弟姐妹/配偶数量):** 大部分乘客没有同乘的兄弟姐妹或配偶 (SibSp=0)。箱线图显示有一些乘客拥有 3 个或更多同乘的亲属，最高达到 8 个。这些数量较大的家庭或团体在图上显示为异常值。
*   **Parch (同乘的父母/子女数量):** 大部分乘客没有同乘的父母或子女 (Parch=0)。箱线图显示有一些乘客拥有 3 个或更多同乘的亲属，最高达到 6 个。与 SibSp 类似，这些较大的家庭或团体在图上显示为异常值。



**异常值处理方法讨论：**

对于这些识别出的潜在异常值，我们需要根据它们可能代表的含义以及对后续分析或模型的影响来决定如何处理。

1.  **Age (年龄):** 较高的年龄值（例如，70岁以上）在现实中是存在的，并非一定是录入错误。如果我们的分析或模型对年龄非常敏感，或者需要更精细地处理极端年龄，可以考虑：
    *   **保留:** 如果这些年龄值是真实的，保留它们可以反映数据集的真实分布。
    *   **转换:** 对年龄进行分组（例如，儿童、青少年、成年、老年）可能比处理具体数值更能捕捉其与生存率的关系，并减少极端值的影响。
    *   **封顶 (Capping):** 将高于某个阈值（例如，数据的 95% 或 99% 分位数）的年龄值设置为该阈值，以减轻极端高值的影响。

2.  **Fare (船票价格):** 极端高的船票价格很可能对应于头等舱乘客。这些高价格虽然在统计上是异常值，但在业务上却是有意义的，代表了数据集中的一个重要群体。
    *   **保留:** 保留这些高票价可以反映不同船舱等级的价格差异，这与生存率可能高度相关。删除它们会丢失重要的信息。
    *   **转换:** 船票价格的分布呈严重的右偏（长尾），对数转换（例如，`np.log1p(df['Fare'])`）是处理这种偏态分布和减轻极端高值影响的常用方法。转换后的数据分布更接近正态，有助于某些假设依赖于正态性的模型。
    *   **基于业务理解处理:** 可以结合 'Pclass' 列来理解票价，例如，分析不同 Pclass 的票价分布。

3.  **SibSp 和 Parch (同乘亲属数量):** 较大的家庭或团体人数在统计上是异常值，但它们代表了乘客的家庭结构，这可能与生存率有关（例如，与家人一起旅行可能影响行为或救援优先级）。
    *   **保留:** 保留这些值可以反映家庭规模的影响。
    *   **合并或转换:** 可以将 SibSp 和 Parch 合并为一个新的特征 'Family_Size' (SibSp + Parch + 1)，然后分析家庭大小与生存率的关系。对于家庭大小，也可以考虑进行分组（例如，独自旅行、小家庭、大家庭）。
    *   **封顶:** 如果极端大的家庭人数（例如，8个SibSp 或 6个Parch）被认为是异常情况或录入错误，可以考虑封顶。但在这个数据集中，这些高值可能代表真实情况。


**总结处理策略（在此步骤只讨论，不执行具体修改）：**

*   **Age:** 考虑保留或进行适当的分组/转换，具体取决于后续分析的需求。中位数填充已在处理缺失值时完成，这已经是一种对分布影响较小的处理方式。
*   **Fare:** **强烈建议保留**这些高票价，因为它们代表了重要的群体特征（如船舱等级）。为了减轻其对某些模型的潜在影响，**对 'Fare' 进行对数转换**是常用的且合理的方法，这将在特征工程步骤中进行。
*   **SibSp 和 Parch:** 考虑**保留**这些值，并将它们合并为 'Family_Size' 特征进行分析，或者对家庭大小进行分组。


**重要提示:**

在本阶段，我们主要关注异常值的识别和讨论。具体的处理方法（如转换或封顶）将在后续的特征工程或数据预处理步骤中根据具体建模需求进行。对于 'Fare' 的对数转换和 SibSp/Parch 的合并，是基于经验和业务理解的常用处理方式，将在后续步骤实现。对于 Age 的极端值，目前的中位数填充已经处理了缺失值，如果需要进一步处理极端值，可以在后续步骤考虑分组或封顶。

## 双变量分析

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 基于 'SibSp' 和 'Parch' 创建 'Family_Size'
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1

# 2. 基于 'Family_Size' 创建 'Family_Type'
# 根据 Family_Size 的分布，我们可以大致划分为：独自旅行 (1人)，小家庭 (2-4人)，大家庭 (>4人)
df['Family_Type'] = 'Solo'
df.loc[(df['Family_Size'] > 1) & (df['Family_Size'] <= 4), 'Family_Type'] = 'Small'
df.loc[df['Family_Size'] > 4, 'Family_Type'] = 'Large'

# 3. 基于 'Fare' 列的分布，创建 'Fare_Category'
# 使用分位数进行划分 (例如，四等分)
df['Fare_Category'] = pd.qcut(df['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very_High'])

# 4. 基于 'Age' 列的分布，创建 'Age_Group'
# 根据业务知识或数据分布进行划分
# 例如：儿童 (0-12), 青少年 (13-18), 青年 (19-35), 中年 (36-60), 老年 (>60)
bins = [0, 12, 18, 35, 60, df['Age'].max()]
labels = ['Child', 'Teenager', 'Young_Adult', 'Adult', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# 5. 查看新创建的特征的分布和与生存率的关系
categorical_new_features = ['Family_Size', 'Family_Type', 'Fare_Category', 'Age_Group']

plt.figure(figsize=(20, 5))
plot_index = 1
for col in categorical_new_features:
    plt.subplot(1, 4, plot_index)
    sns.countplot(data=df, x=col, hue='Survived', palette='viridis')
    plt.title(f'Survival Count by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.legend(title='Survived', loc='upper right')
    plot_index += 1
plt.tight_layout()
plt.show()

# Display the first few rows with new features
print("\nDataset with New Features:")
display(df.head())

```

## 总结

### EDA 主要发现回顾

到目前为止，我们对 Titanic 数据集进行了全面的探索性数据分析（EDA），主要发现了以下关键信息：

1.  **数据集概览和缺失值:** 数据集包含了 891 名乘客的信息。我们在 'Age'、'Cabin' 和 'Embarked' 列发现了缺失值。其中 'Cabin' 的缺失率极高（77%），'Age' 缺失率接近 20%，而 'Embarked' 只有少量缺失。我们对 'Age' 进行了中位数填充，对 'Embarked' 进行了众数填充，并将 'Cabin' 转换为了一个二元特征 'Has_Cabin'。
2.  **单变量分析:**
    *   数值型变量如 'Age' 和 'Fare' 的分布存在偏态，尤其是 'Fare' 严重右偏。
    *   分类型变量如 'Survived', 'Pclass', 'Sex', 'Embarked' 的频率分布揭示了不同类别的人数比例。例如，未生存人数多于生存人数，男性乘客远多于女性，三等舱乘客最多。
3.  **双变量和多变量分析:**
    *   **生存率与性别:** 女性的生存率远高于男性。
    *   **生存率与船票等级 (Pclass):** 头等舱乘客的生存率远高于二等舱和三等舱乘客，船票等级是影响生存的重要因素。
    *   **生存率与船票价格 (Fare):** 票价越高，生存率越高，这与船票等级的发现一致。
    *   **生存率与登船港口 (Embarked):** 不同登船港口的生存率存在差异。
    *   **生存率与是否有船舱信息 (Has_Cabin):** 有船舱信息的乘客生存率更高，这可能与船票等级相关（高等级船舱信息更完整）。
    *   **变量间的相关性:** 'SibSp' 和 'Parch' 之间存在正相关；'Fare' 与 'Pclass' 存在显著负相关（票价越高，等级越低），与 'Survived' 和 'Has_Cabin' 存在正相关。
4.  **特征工程:** 基于 EDA 的发现，我们创建了新的特征：
    *   **Family_Size (家庭大小):** SibSp + Parch + 1，反映乘客是否独自旅行或家庭规模。
    *   **Family_Type (家庭类型):** 将 Family_Size 划分为 Solo, Small, Large。发现小家庭生存率较高。
    *   **Fare_Category (票价类别):** 将 Fare 划分为 Low, Medium, High, Very_High。高票价类别生存率更高。
    *   **Age_Group (年龄分组):** 将 Age 划分为 Child, Teenager, Young_Adult, Adult, Senior。发现儿童和青年生存率相对较高。
    *   **Log_Fare (对数转换票价):** 对偏态分布的 Fare 进行了对数转换，使其分布更接近正态。


### EDA 发现对后续步骤的指导意义

EDA 阶段获得的这些深刻洞察将直接指导我们后续的数据科学流程：

1.  **数据预处理:**
    *   **缺失值处理:** EDA 帮助我们识别了缺失值的位置和比例，并基于此选择了合适的填充或转换策略 ('Age', 'Embarked' 填充，'Cabin' 转换为二元特征)。后续可以直接使用处理后的数据。
    *   **异常值处理:** EDA 通过箱线图等发现了 'Fare', 'SibSp', 'Parch' 中的异常值。我们讨论了这些异常值的意义，并决定对 'Fare' 进行对数转换，这有助于减少极端值对模型的影响，并在特征工程中已经实现。对于 'SibSp' 和 'Parch' 的极端值，结合 Family_Size/Type 特征进行分析是更合适的处理方式。
    *   **数据转换:** EDA 显示 'Fare' 呈偏态分布，指导我们进行了对数转换 ('Log_Fare')，这将有助于依赖于数据正态性假设的某些模型。

2.  **特征选择:**
    *   EDA 发现了与生存率高度相关的特征，如 'Sex', 'Pclass', 'Fare'/'Log_Fare', 'Has_Cabin' 以及新创建的 'Family_Type', 'Fare_Category', 'Age_Group'。这些特征是构建预测模型时的重要候选特征。
    *   对于相关性较高的特征对（如 SibSp 和 Parch，或 Pclass 和 Fare），我们可以考虑保留其中一个或使用它们的组合特征 ('Family_Size', 'Fare_Category')，避免多重共线性问题，同时捕捉更复杂的模式。

3.  **模型选择:**
    *   EDA 揭示了特征与生存率之间存在的非线性关系（例如，Family_Size 的中间部分生存率较高）。这提示我们，线性模型可能不足以完全捕捉这些关系，可以考虑更复杂的非线性模型，如决策树、随机森林或梯度提升模型。
    *   二元分类问题（生存/未生存）的性质决定了我们需要选择适合分类任务的模型。

4.  **模型解释:**
    *   通过 EDA 了解到的特征重要性（如性别、船票等级对生存率的影响）将帮助我们解释模型的预测结果，例如，为什么模型预测某个乘客生存或未生存，很可能是因为他们的性别、船票等级等特征。
    *   新创建的特征（如 Family_Type, Age_Group）使得模型解释更具业务意义。



### EDA 的重要性再强调

本次 EDA 过程充分展示了其在数据科学项目中的核心价值：

*   **避免盲目性:** 在没有对数据进行充分探索之前直接建模，很可能因为不了解数据的特性（如缺失值、异常值、偏态分布、重要关联）而导致模型性能不佳或产生误导性结论。
*   **提供方向:** EDA 帮助我们识别哪些特征是重要的，哪些模式是存在的，从而为数据清洗、特征工程和模型选择提供了明确的方向和依据。
*   **生成假设:** EDA 过程中的发现会激发关于数据背后机制的假设（例如，为什么女性生存率高？），这些假设可以在后续建模中进一步验证。
*   **提高效率和效果:** 通过在早期阶段发现并解决数据问题，理解关键特征，可以避免在建模阶段走弯路，提高整个项目的效率和最终模型的预测效果。
